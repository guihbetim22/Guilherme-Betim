```python
# Importação das devidas bibliotecas que seram utilizadas

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Realizando a leitura do arquivo csv 

dados = pd.read_csv("vgsales.csv")
df = pd.read_csv("vgsales.csv")
```

```python
# criando um dataframe com os dados
df_games = df[['Name', 'Global_Sales', 'Publisher']].sort_values('Global_Sales',ascending=False).head(10)

# definindo a cor
sns.set_theme(style="white", palette="pastel")

# definir o título e as devidas infos no grafico
sns.barplot(data=df_games, x='Global_Sales', y='Name', orient='h').set(title='Top 10 Games nas vendas mundiais')
plt.xlabel("Vendas mundiais")
```

```python
## selecionando as vendas por regiões (Continente e País)

sales = ['NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales']

## criando o grafico com a bliblioteca Plotly
fig = go.Figure()

## definindo a escala(tamanho) do grafico
fig.update_layout(height=650, width=650)

## traçando os dados(anos) com os paises dentro do grafico
fig.add_trace(go.Scatter(x=dados["Year"], y=dados["NA_Sales"], mode="markers", name="NA Sales"))
fig.add_trace(go.Scatter(x=dados["Year"], y=dados["EU_Sales"], mode="markers", name="EU Sales"))
fig.add_trace(go.Scatter(x=dados["Year"], y=dados["JP_Sales"], mode="markers", name="JP Sales"))
fig.add_trace(go.Scatter(x=dados["Year"], y=dados["Other_Sales"], mode="markers", name="Other Sales"))

## definir o título e as devidas infos no grafico
fig.update_layout(title="Vendas por ano", xaxis_title="Ano", yaxis_title="Vendas")

## exibir o grafico
fig.show()
```

```python
## definindo a cor do grafico
sns.set_theme(style="white", palette="pastel")

## criando um dataframe com os dados
df_sales = df[['Year', 'Global_Sales', 'Genre']].groupby(['Year', 'Genre']).sum()

## definir o título e as devidas infos
ax = sns.lineplot(data=df_sales, x='Year', y= 'Global_Sales', hue = 'Genre').set(title='Vendas mundiais por genero')

## definindo o tamanho do grafico
plt.figure(figsize=(12,8))
```

```python
## criando um grafico de pizza e difinindo o titulo
fig = px.pie(names=dados.Genre.value_counts().index, values=dados.Genre.value_counts().values, title='Distribuição de Games mais comprados por genero')

## definindo o tamanho do grafico
fig.update_layout(height=700, width=700, 

## definindo a escala da legenda                 
legend=dict(x = 1.1, y = 0.01))

## exibir o grafico
fig.show()
```

```python
## definindo a cor externa do grafico e o tamando do mesmo
plt.figure(facecolor = "darkgray",figsize = (18, 6))
## legenda
plt.xticks(rotation=90)

##plot do grafico 
ax= sns.barplot(x = dados.Platform.value_counts().index, y = dados.Platform.value_counts().values)

## definir o título e as devidas infos no grafico
ax.set(xlabel='Plataformas',ylabel='Contagem', title='Consoles mais utilizados')

## definindo a cor interna do grafico
ax.set_facecolor("white")
ax.bar_label(ax.containers[0])

## exibir o grafico
plt.show()
```

```python
## criando um dataframe com os dados
Publishers = pd.DataFrame(df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending = False).head(10))

## definir o título e o tamanho da grafico
Publishers['Global_Sales'].plot(kind='barh', figsize=(6,4),title= "Top 10 Desenvolvedoras de Games Mundial")
```
